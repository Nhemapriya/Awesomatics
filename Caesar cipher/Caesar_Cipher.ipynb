```
Symbols = 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz'
key_size = len(Symbols)
```

```
def getMode():
  while True:
    print('You want to encrypt or decrypt the message ?')
    mode = input().lower()
    mode_array = ['encrypt','e','decrypt','d']
    if mode in mode_array:
      return mode
    else:
      print("Enter option as : encrypt or decrypt or e or d")
```

```
def getMessage():
  print('Enter your message information')
  return input()
```

```
def getKey():
  key = 0
  while True:
    print('Enter the key number (1-%s)' % (key_size))
    key = int(input())
    if (key >= 1 and key <= key_size):
      return key
```

```
def ProcessedMessage(mode, message , key):
  if mode[0] == 'd':
    key = -key
  processed = ''

  for symbol in message:
    symbol_index = Symbols.find(symbol)
    #if the symbol is not found , include it without any processing
    if symbol_index == -1:
      processed+=symbol
    else:
      #perform encryption or decryption operation
      symbol_index +=key 

      if symbol_index >= len(Symbols):
        symbol_index -= len(Symbols)
      elif symbol_index < 0:
        symbol_index += len(Symbols)

      processed+=Symbols[symbol_index]
  return processed
```

```
# Implementing Caesar Cipher 
#Function Calls

#ENCRYPTION
mode = getMode()
message = getMessage()
key = getKey()
print('Your processed text is :')
print(ProcessedMessage(mode , message , key))
```
